```python


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python


from google.colab import files

uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))
  # Then you can read it into a pandas dataframe:
  df = pd.read_excel(fn)
  # Now you can work with your dataframe!
```

```python


df['idade'].describe()
```

```python


df['tipo'].value_counts().plot(kind='pie', figsize=(15, 6), autopct='%1.1f%%')
plt.title('Tipo')
plt.ylabel('')
plt.show()
```

```python


df['municipio'].value_counts().plot(kind='barh', figsize=(15, 6))
plt.title('Municipio')
plt.gca().invert_yaxis()
plt.show()
```

```python


df['origem_chamado'].value_counts().plot(kind='bar', figsize=(15, 6))
plt.title('Origem do Chamado')
plt.ylabel('Quantidade')
plt.xlabel('Origem')
plt.show()
```

```python


df['sexo'] = df['sexo'].str.upper()
```

```python


df['sexo'].value_counts(normalize=True).plot(kind='bar', figsize=(15, 6))
plt.title('Solicitações por Sexo')
plt.ylabel('Percentual')
plt.xlabel('Sexo')
plt.show()
```

```python


# Create a new column for age groups
df['faixa_etaria'] = pd.cut(df['idade'], bins=np.arange(0, 100, 10), labels=np.arange(10, 100, 10))

# Group the data by sex and age group
grouped = df.groupby(['sexo', 'faixa_etaria']).size().unstack()

# Create the stacked bar chart
ax = grouped.plot(kind='bar', stacked=True, figsize=(15, 6))

# Add labels and title
plt.title('Pirâmide Etária por Sexo')
plt.ylabel('Quantidade')
plt.xlabel('Faixa Etária')

# Show the plot
plt.show()
```

```python


df.groupby('motivo_desfecho')['ID'].count().plot.barh(figsize=(15, 6))
plt.title('Motivo do Desfecho')
plt.gca().invert_yaxis()
plt.show()
```

```python


df.groupby('tipo')['ID'].count().plot(kind='bar', figsize=(15, 6))
plt.title('Quantidade de Solicitações por Tipo')
plt.ylabel('Quantidade')
plt.xlabel('Tipo')
plt.show()
```
